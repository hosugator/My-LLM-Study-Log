# 미국 주식 분석하기

```python
import yfinance as yf
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# 오늘 날짜를 구합니다.
end_date = datetime.now().date()

# period of days
day_period = 30

# 30일 전 날짜를 구합니다.
start_date = end_date - timedelta(days=day_period)

# 엔비디아(NVIDIA) 종목 코드를 지정합니다.
nvidia_ticker = "NVDA"

# 지정된 기간 동안의 주가 데이터를 다운로드합니다.
# 자동으로 시가(Open), 고가(High), 저가(Low), 종가(Close), 거래량(Volume) 데이터가 포함됩니다.
df = yf.download(nvidia_ticker, start=start_date, end=end_date)

# 시각화를 위해 새로운 그림(figure)을 만듭니다.
plt.figure(figsize=(12, 7))

# 종가, 고가, 저가를 각각 그래프로 그립니다.
plt.plot(df.index, df["Close"], label="Close Price", color="blue")
plt.plot(df.index, df["High"], label="High Price", color="green")
plt.plot(df.index, df["Low"], label="Low Price", color="red")

# 그래프에 제목과 축 레이블을 추가합니다.
plt.title(f"NVIDIA (NVDA) Stock Prices ({start_date} ~ {end_date})", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)

# 그래프를 화면에 표시합니다.
plt.show()
```
