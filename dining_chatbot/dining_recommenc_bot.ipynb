```python
import requests


def get_url_content(url):
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            return response.text
        else:
            response.raise_for_status()

    except requests.exceptions.RequestException as e:
        print(f"Request Exception: {e}")
        return None

# url을 통해 API 호출하기
result = get_url_content(url=url)  # ... code here

if result is not None:

    print(result)
else:
    print("Failed to retrieve URL content.")
```

```python
import json

data = json.loads(result)
restaurants = data['documents']
restaurants
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
openai_api_key = os.getenv("OPENAI_API_KEY")  # OpenAI API 키

# 2. OpenAI 클라이언트 초기화
client = OpenAI()
```

```python
def extract_keyword_from_question(question, model="gpt-3.5-turbo"):
    """
    사용자 입력으로부터 키워드를 뽑아낸다
    """
    response = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "system",
                "content": "다음 질문에서 가장 중요한 키워드 단어 하나만 뽑아줘."
            },
            {
                "role": "user",
                "content": question
            }
        ],
        temperature=0,
        max_tokens=100,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )
    return response.choices[0].message.content.strip()
```

```python
question = "초밥 맛있는 곳"

keyword = extract_keyword_from_question(question)
print(f"key word is: {keyword}")

encode_keyword = quote(keyword)
```

```python
import json

# 이 코드는 이전 단계에서 result 변수에 API 응답이 담겨 있다고 가정합니다.
# result = get_url_content(url=url)
# data = json.loads(result)
# restaurants = data['documents']

# 1. 원하는 키만 추출하여 새로운 리스트 생성
filtered_restaurants = []
for restaurant in restaurants:
    filtered_data = {
        "이름": restaurant.get('place_name'),
        "도로명주소": restaurant.get('road_address_name'),
        "카테고리": restaurant.get('category_name')
        # 필요한 다른 키도 추가할 수 있습니다.
    }
    filtered_restaurants.append(filtered_data)

# 2. 파이썬 리스트를 JSON 문자열로 변환
# ensure_ascii=False로 한글 깨짐 방지
# indent=2로 가독성 높이기
json_data_for_gpt = json.dumps(filtered_restaurants, ensure_ascii=False, indent=2)

# 3. GPT에 보낼 프롬프트 구성
prompt = f"""
다음은 검색된 음식점 목록입니다. JSON 형식으로 제공됩니다:

{json_data_for_gpt}

{keyword}에 가장 잘 맞는 음식점 3곳을 추천하고,
각각의 식당에 대해 짧은 평론을 한두 문장으로 작성해줘.
"""

# 4. GPT-3.5-turbo에 단일 요청 전송
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {
            "role": "system",
            "content": "음식점 추천 전문가입니다. 제공된 JSON 데이터를 바탕으로 최고의 식당을 선별하고 평론해 드립니다."
        },
        {
            "role": "user",
            "content": prompt
        }
    ]
)

# 5. 결과 출력
recommendations = response.choices[0].message.content.strip()
print(recommendations)
```
